```python
#Pregunta a analizar:
# ¿Existe una diferencia significativa de temperaturas entre Salto y Treinta y Tres en el mes de Enero 2023? ¿Cuál deberia ser mi elegido para vacacionar segun mis preferencias?

#Las bibliotecas necesarias para crear mi codigo:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Leer los datasets de temperaturas para Salto y Treinta y Tres:
temperaturasalto= pd.read_csv("ppt_tmax_tmin_sg_2023.csv")
temperaturatreintaytres= pd.read_csv("ppt_tmax_tmin_tyt_2023.csv")

#Filtrar los datos unicamente del mes de enero
temperaturassaltoenero=temperaturasalto[(temperaturasalto["fecha"]>="2023-01-01")&(temperaturasalto["fecha"]<="2023-01-31")]
temperaturastreintaytresenero=temperaturatreintaytres[(temperaturatreintaytres["fecha"]>="2023-01-01")&(temperaturatreintaytres["fecha"]<="2023-01-31")]


#En el primer subplot hallare la temperatura máxima de cada departamento
plt.subplot(2,1,1)
plt.plot(temperaturassaltoenero["fecha"],temperaturassaltoenero["temperaturaMaxima"],label="salto",color="red")
plt.plot(temperaturastreintaytresenero["fecha"],temperaturastreintaytresenero["temperaturaMaxima"],label="treintaytres",color="green")
plt.title("Temperatura Máxima")
plt.xlabel("Fecha")
plt.ylabel("Temp. máxima(°C)")
plt.xticks(range(len(temperaturassaltoenero['fecha'])), temperaturassaltoenero['fecha'], rotation=45, fontsize=6, ha='right')
plt.legend()
plt.tight_layout()

#En el segundo subplot hallare la temperatura minima de cada departamento
plt.subplot(2,1,2)
plt.plot(temperaturassaltoenero["fecha"],temperaturassaltoenero["temperaturaMinima"],label="salto",color="red")
plt.plot(temperaturastreintaytresenero["fecha"],temperaturastreintaytresenero["temperaturaMinima"],label="treintaytres",color="green")
plt.title("Temperatura Minima")
plt.xlabel("Fecha")
plt.ylabel("Temp. minima(°C)")
plt.xticks(range(len(temperaturassaltoenero['fecha'])), temperaturassaltoenero['fecha'], rotation=45, fontsize=6, ha='right')
plt.legend()
plt.tight_layout()

#Mostrar el gráfico
plt.show()
```

Las temperaturas en Salto y Treinta y Tres muestran variabilidad a lo largo del mes de enero de 2023. Se observa una tendencia a temperaturas mas elevadas en el departamento de Salto. Las temperaturas minimas de Salto mayormente se encuentran sobre los 20°C y en Treinta y Tres las temperaturas minimas se encuentran por debajo de los 20°C, teniendo una media sobre los 15°C.
 Reflejando un clima generalmente más cálido en comparación con Treinta y Tres. Este patrón se mantiene durante todo el mes de enero, con temperaturas más elevadas en Salto.
